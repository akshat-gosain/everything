```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility (optional)
np.random.seed(42)

# Generate 1000 random values from a uniform distribution
random_values = np.random.uniform(size=1000)

# Create an array of time points (assuming a unit time interval)
time_points = np.arange(1, 1001)

# Plot the random values over time
plt.plot(time_points, random_values, label='Random Values', color='blue')
plt.title('Random Values Over Time')
plt.xlabel('Time')
plt.ylabel('Random Value')
plt.legend()
plt.show()

# Save the generated data in .npy format
np.save('random_values.npy', random_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def skew_tent_map(x, a):
    if x < a:
        return x / a
    else:
        return (1 - x) / (1 - a)

# Parameters
a = 0.499  # Skew parameter

# Number of iterations
iterations = 1000

# Initial value
x_values = [0.37]

# Generate values using the skew tent map
for _ in range(iterations):
    x_next = skew_tent_map(x_values[-1], a)
    x_values.append(x_next)

# Plot the skew tent map values over time
plt.plot(x_values[:-1], x_values[1:], marker='o', linestyle='', markersize=2, label='Skew Tent Map')
plt.title('Skew Tent Map')
plt.xlabel('$x_n$')
plt.ylabel('$x_{n+1}$')
plt.legend()
plt.show()
np.save('dataset_2.npy', x_values)
```
