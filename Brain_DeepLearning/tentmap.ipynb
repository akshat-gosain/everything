```python
import matplotlib.pyplot as plt
import numpy as np

# Define the skew tent map function
def skew_tent_map(x, b):
    if 0 <= x <= b:
        return x / b
    elif b < x <= 1:
        return (1 - x) / (1 - b)

# Initial value of x
x = 0.0001

# Value of b
b = 0.5  # Change this to your desired value of b

# Number of iterations to perform
num_iterations = 1000

# Store the x values at each iteration
x_values = [x]

for _ in range(num_iterations):
    x = skew_tent_map(x, b)
    x_values.append(x)

# Plot the first return map
plt.plot(x_values[:-1], x_values[1:], 'o')
plt.xlabel('x[n]')
plt.ylabel('x[n+1]')
plt.title('First Return Map for Skew Tent Map')
plt.show()
```
