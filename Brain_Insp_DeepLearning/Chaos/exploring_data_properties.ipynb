```python
import numpy as np
import matplotlib.pyplot as plt
# Load the dataset
file_path = '/content/drive/MyDrive/BITS-Courses/BIDL/lecture 9-10 chaos/dataset_1.npy'  # Make sure to adjust the filename if needed
data_1 = np.load(file_path)

# Load the dataset
file_path_2 = '/content/drive/MyDrive/BITS-Courses/BIDL/lecture 9-10 chaos/dataset_2.npy'  # Make sure to adjust the filename if needed
data_2 = np.load(file_path_2)
```

# Dataset 1

```python
##
time_points = np.arange(0, 1000)

plt.plot(time_points[0:100], data_1[0:100], label='Dataset_1', color='blue')
plt.title('Dataset 1 Over Time')
plt.xlabel('Time')
plt.ylabel('Data 1 Value')
plt.legend()
plt.show()
```

# Dataset 2

```python
##
time_points = np.arange(0, 1000)

plt.plot(time_points[0:100], data_2[0:100], label='Dataset_2', color='blue')
plt.title('Dataset 2 Over Time')
plt.xlabel('Time')
plt.ylabel('Data 2 Value')
plt.legend()
plt.show()
```

# Histogram of dataset 1

```python
# Plot the histogram
plt.hist(data_1, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Dataset 1')
plt.xlabel('Data 1 Value')
plt.ylabel('Frequency')
plt.show()
```

# Histogram of Dataset 2

```python
# Plot the histogram
plt.hist(data_2, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Dataset 2')
plt.xlabel('Data 2 Value')
plt.ylabel('Frequency')
plt.show()
```

# First Return Map of Dataset 1

```python
# Plot the skew tent map values over time
plt.plot(data_1[:-1], data_1[1:], marker='o', linestyle='', markersize=2, label='Data 1')
plt.title('First Return of Dataset 1')
plt.xlabel('$x_n$')
plt.ylabel('$x_{n+1}$')
plt.legend()
plt.show()
```
